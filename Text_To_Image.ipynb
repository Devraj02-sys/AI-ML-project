```python
import torch
import gradio as gr
from diffusers import StableDiffusionPipeline

# Load model
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# Function to generate image
def generate_image(prompt, steps=20):
    image = pipe(prompt, num_inference_steps=steps, guidance_scale=7.5).images[0]
    return image

# Gradio UI
interface = gr.Interface(
    fn=generate_image,
    inputs=[gr.Textbox(label="Enter Prompt"), gr.Slider(1, 50, value=20, label="Inference Steps")],
    outputs="image",
    title="AI Text-to-Image Generator",
    description="Enter a description and generate an AI image.",
)

# Run app
interface.launch()
```
